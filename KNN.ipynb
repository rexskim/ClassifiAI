```python
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.neighbors import KNeighborsClassifier

seed = 1234
```

```python
X, y = fetch_openml('cifar_10', as_frame=False, return_X_y=True)
```

```python
X, X_te, y, y_te = train_test_split(X, y, test_size=0.2, random_state=seed, shuffle=True)
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X_te = scaler.transform(X_te)
X_tr, X_val, y_tr, y_val = train_test_split(X, y, test_size=0.25, random_state=seed, shuffle=True)
```

```python
figure, axes = plt.subplots(1, figsize=(6, 6))
k_vals = [1, 5, 10, 50, 100, 110]
scores = []
tscores = []
for k in k_vals:
 knn = KNeighborsClassifier(n_neighbors=k)
 knn.fit(X_tr, y_tr)
 tr_pred = knn.predict(X_tr)
 y_preds = knn.predict(X_te)
 score = sklearn.metrics.accuracy_score(y_te, y_preds)
 tscore = sklearn.metrics.accuracy_score(y_tr, tr_pred)
 scores.append(1-score)
 tscores.append(1-tscore)
axes.semilogx(k_vals, scores, color='green', label="Test Error")
axes.semilogx(k_vals, tscores, color='red', label="Training Error")
axes.set_title("kNN: Error as k Increases")
axes.set_xlabel("# of kNN")
axes.set_ylabel("Error")
axes.legend()

```

```python
axes.set_title("kNN: Error as k Increases")
axes.set_ylabel("Accuracy")
axes.legend()
axes.imshow()
```
