```
import numpy as np
import matplotlib.pyplot as plt

import sklearn
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.inspection import DecisionBoundaryDisplay

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree

import tensorflow as tf
from tensorflow.keras import datasets, layers, models

import warnings
warnings.filterwarnings('ignore')

# Fix the random seed for reproducibility
# !! Important !! : do not change this
seed = 1234
np.random.seed(seed) 
```

```
X, y = fetch_openml('cifar_10', as_frame=False, return_X_y=True)
```

```
X, X_te, y, y_te = train_test_split(X, y, test_size=0.2, random_state=seed, shuffle=True)
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X_te = scaler.transform(X_te)
X_tr, X_val, y_tr, y_val = train_test_split(X, y, test_size=0.25, random_state=seed, shuffle=True)
```

```
# Jon is building out feedforward neural network

y = y.astype(int)

import random
import json
random.seed(seed)
learning_rates = [0.0005, 0.005]
widths = [8, 128, 512]
depths = [1,3,5]
regularization_strengths = [0.0005, 0.005]
activations = ['logistic', 'relu']
batch_sizes = ['auto', 32, 128, 512]
configurations = []
train_accuracy = []
val_accuracy = []
test_accuracy = []
n = 30 # Number of classifiers to be created
# Fit and score all the classifiers
for i in range(n):
    config = {"learning_rate_init":random.choice(learning_rates), \
        "hidden_layer_sizes":tuple([random.choice(widths) for i in range(random.choice(depths))]), \
        "alpha":random.choice(regularization_strengths), \
        "activation":random.choice(activations), \
        "batch_size":random.choice(batch_sizes)}

    configurations.append(config)
    classifier = MLPClassifier(learning_rate_init=config["learning_rate_init"], \
        hidden_layer_sizes=config["hidden_layer_sizes"], \
        alpha=config["alpha"], \
        activation=config["activation"], \
        batch_size=config["batch_size"], \
        learning_rate='adaptive', \
        random_state=seed)

    classifier.fit(X_tr, y_tr)
    train_accuracy.append(classifier.score(X_tr, y_tr))
    val_accuracy.append(classifier.score(X_val, y_val))
    test_accuracy.append(classifier.score(X_te, y_te))
# Determine Best Hyperparameters Based on Validation Data
best_hyperparameters_i = val_accuracy.index(max(val_accuracy))
# Create the table
col_sizes = [6, 19, 22, 7, 11, 11, 11, 9, 10]
table = [[""]*(9) for i in range(n + 1)]
table[0][0] = "Run #"
table[0][1] = "Learning Rate Init"
table[0][2] = "Hidden Layer Sizes"
table[0][3] = "Alpha"
table[0][4] = "Activation"
table[0][5] = "Batch Size"
table[0][6] = "Train Acc"
table[0][7] = "Val Acc"
table[0][8] = "Test Acc"
for i in range(n):
    t_i = i + 1
    table[t_i][0] = str(t_i)
    table[t_i][1] = str(configurations[i]["learning_rate_init"])
    table[t_i][2] = str(configurations[i]["hidden_layer_sizes"])
    table[t_i][3] = str(configurations[i]["alpha"])
    table[t_i][4] = str(configurations[i]["activation"])
    table[t_i][5] = str(configurations[i]["batch_size"])
    table[t_i][6] = str("{0:<6}".format("%.6s"%train_accuracy[i]) + " ")
    table[t_i][7] = str("{0:<6}".format("%.6s"%val_accuracy[i]) + " ")
    if i == best_hyperparameters_i:
        table[t_i][8] = str("{0:<6}".format("%.6s"%test_accuracy[i]) + " ")
# Print the table
for j, row in enumerate(table):
    for i, cell in enumerate(row):
        print(("{0:<" + str(col_sizes[i]) + "}").format(cell) + " ", end="")
    if j - 1 == best_hyperparameters_i:
        print("⬅️", end="")
    print()

# can show how acc changes as we change any individual hyper parameter
```
