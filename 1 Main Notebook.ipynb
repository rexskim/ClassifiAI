```
# Accuracy vs Regularization strength below
coefs_ = []
train_accuracies = []
test_accuracies = []
cs = [0, 0.1, 1, 10, 50]
for c in cs:
    if c == 0:
        classifier = LogisticRegression(penalty='none', fit_intercept=True)
    else:
        classifier = LogisticRegression(penalty='l1', C=c, fit_intercept=True, solver='liblinear')
    classifier.fit(X_tr, y_tr)
    coefs_.append(classifier.coef_.ravel().copy())
        
    train_predictions = classifier.predict(X_tr)	 # Training
    train_accuracy = accuracy_score(y_tr, train_predictions)
    train_accuracies.append(train_accuracy)
        
    # Compute and print accuracy and AUC on the test data
    test_predictions = classifier.predict(X_te)
    test_accuracy = accuracy_score(y_te, test_predictions) 
    test_accuracies.append(test_accuracy)

        
    
fig, axes = plt.subplots()
axes.semilogx(cs, train_accuracies, color='red', label='training accuracy')
axes.semilogx(cs, test_accuracies, color='blue', label='testing accuracy')
    
axes.set_xlabel('regulaization strength', fontsize=14)
axes.set_ylabel('accuracy', fontsize=14)
    
axes.legend()
```

```
    # Jon is building out feedforward neural network

y = y.astype(int)

import random
import json
random.seed(seed)
learning_rates = [0.0005, 0.005]
widths = [8, 128, 512]
depths = [1,3,5]
regularization_strengths = [0.0005, 0.005]
activations = ['logistic', 'relu']
batch_sizes = ['auto', 32, 128, 512]
configurations = []
train_accuracy = []
val_accuracy = []
test_accuracy = []
n = 5 # Number of classifiers to be created
# Fit and score all the classifiers
for i in range(n):
    config = {"learning_rate_init":random.choice(learning_rates), \
        "hidden_layer_sizes":tuple([random.choice(widths) for i in range(random.choice(depths))]), \
        "alpha":random.choice(regularization_strengths), \
        "activation":random.choice(activations), \
        "batch_size":random.choice(batch_sizes)}

    configurations.append(config)
    classifier = MLPClassifier(learning_rate_init=config["learning_rate_init"], \
        hidden_layer_sizes=config["hidden_layer_sizes"], \
        alpha=config["alpha"], \
        activation=config["activation"], \
        batch_size=config["batch_size"], \
        learning_rate='adaptive', \
        random_state=seed)

    classifier.fit(X_tr, y_tr)
    train_accuracy.append(classifier.score(X_tr, y_tr))
    val_accuracy.append(classifier.score(X_val, y_val))
    test_accuracy.append(classifier.score(X_te, y_te))
# Determine Best Hyperparameters Based on Validation Data
best_hyperparameters_i = val_accuracy.index(max(val_accuracy))
# Create the table
col_sizes = [6, 19, 22, 7, 11, 11, 11, 9, 10]
table = [[""]*(9) for i in range(n + 1)]
table[0][0] = "Run #"
table[0][1] = "Learning Rate Init"
table[0][2] = "Hidden Layer Sizes"
table[0][3] = "Alpha"
table[0][4] = "Activation"
table[0][5] = "Batch Size"
table[0][6] = "Train Acc"
table[0][7] = "Val Acc"
table[0][8] = "Test Acc"
for i in range(n):
    t_i = i + 1
    table[t_i][0] = str(t_i)
    table[t_i][1] = str(configurations[i]["learning_rate_init"])
    table[t_i][2] = str(configurations[i]["hidden_layer_sizes"])
    table[t_i][3] = str(configurations[i]["alpha"])
    table[t_i][4] = str(configurations[i]["activation"])
    table[t_i][5] = str(configurations[i]["batch_size"])
    table[t_i][6] = str("{0:<6}".format("%.6s"%train_accuracy[i]) + " ")
    table[t_i][7] = str("{0:<6}".format("%.6s"%val_accuracy[i]) + " ")
    if i == best_hyperparameters_i:
        table[t_i][8] = str("{0:<6}".format("%.6s"%test_accuracy[i]) + " ")
# Print the table
for j, row in enumerate(table):
    for i, cell in enumerate(row):
        print(("{0:<" + str(col_sizes[i]) + "}").format(cell) + " ", end="")
    if j - 1 == best_hyperparameters_i:
        print("⬅️", end="")
    print()

# can show how acc changes as we change any individual hyper parameter
```

```
#Cole is building Convolutional nueral network
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))
```

```
y_preds = model.predict(test_images)
y_preds = [np.argmax(y_pred)for y_pred in y_preds]
print(y_preds[0:5])
print(test_labels[0:5])
sklearn_cm = confusion_matrix(test_labels, y_preds)
sklearn_disp = ConfusionMatrixDisplay(confusion_matrix = sklearn_cm)
sklearn_disp.plot()
```
